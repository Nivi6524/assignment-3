```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("athlete_events.csv")
```

```python
data.head()
```

```python
#merging the other dataset containing information about athletes of all years where they participated 
region = pd.read_csv("datasets_31029_40943_noc_regions.csv")
merged_data = pd.merge(data,region,on="NOC",how="left")
```

```python
#merged_data.head()
```

**1)Which sports has higher athelets of age greater than 50 won a gold medalist**


```python
#getting all the gold medalist
gold_medal = merged_data[(merged_data.Medal == 'Gold')]
```

```python
#count of the gold medalist whoes age is greater than 50
print('The no of athletes is',gold_medal['ID'][gold_medal['Age'] > 50].count(), '\n')
```

```python
master = gold_medal['Sport'][gold_medal['Age'] > 50]
plt.figure(figsize=(20, 10))
plt.tight_layout()
sns.countplot(master)
plt.title('Gold Medals for Athletes over age 50')
plt.show()
```

**2)Which age -athletes got the leading gold medalist**

```python
plt.figure(figsize=(20, 10))
plt.title('Distribution of Gold Medals')
sns.countplot(gold_medal['Age'])
plt.show()
```

**3)In which year the count of women gold medalist was high**

```python
womenInOlympics = merged_data[(merged_data.Sex == "F")]
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
sns.countplot(x='Year', data=womenInOlympics)
plt.title('Women medals per edition of the Games')
plt.show()
```

**4)What are the TOP 5 country who won the most medals**

```python
totalGoldMedals = gold_medal.region.value_counts().reset_index(name='Medal').head(5)
g = sns.catplot(x="index", y="Medal", data=totalGoldMedals,
                height=6, kind="bar", palette="muted")
g.despine(left=True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of Medals")
plt.title('TOP 5 country won gold')
plt.show()
```

**5)How was the weight over year for Male Lifters varies**

```python
MenOverTime = merged_data[(merged_data.Sex == 'M')] 
wMenOverTime = MenOverTime.loc[MenOverTime['Sport'] == 'Weightlifting']
plt.figure(figsize=(20, 10))
sns.pointplot('Year', 'Weight', data=wMenOverTime, palette='Set2')
plt.title('Weight over year for Male Lifters')
plt.show()
```
